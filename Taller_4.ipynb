```python
#Pregunta SMART-Ajustada: identificar cuales son las variables que generan mayor impacto en el buro de crédito de Colombia para ayudar a las personas a mejorar su puntaje crediticio.

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Cargar el dataset
df = pd.read_csv('datos_buro_T3.csv')

df.head(5000)
```

```python
# Verificar si hay valores nulos
print(df.isnull().sum())
```

```python
# Ajustamos valores categóricos
df['Género'] = df['Género'].fillna('Desconocido')
```

```python
print(df.isnull().sum())
```

```python
#Codificación de Variables Categóricas

from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Label Encoding para variables binarias
label_encoder = LabelEncoder()
df['Género'] = label_encoder.fit_transform(df['Género'])  # 'Masculino' -> 1, 'Femenino' -> 0
df['Mora_en_pagos'] = label_encoder.fit_transform(df['Mora_en_pagos'])  # 'Sí' -> 1, 'No' -> 0

# One-Hot Encoding para 'Tipo_de_empleo'
df = pd.get_dummies(df, columns=['Tipo_de_empleo'], drop_first=True)  # Evitar colinealidad eliminando una categoría
```

```python
df.head(5000)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Separar las columnas numéricas
numeric_columns = ['Edad', 'Ingreso_mensual', 'Saldo_total', 'Número_de_cuentas',
                   'Historial_crediticio', 'Porcentaje_uso_credito',
                   'Número_de_moras', 'Monto_total_prestado']

# Escalado estándar (media=0, desviación estándar=1)
scaler = StandardScaler()
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])

df.head(5000)
```

```python
# Alternativamente, usamos MinMaxScaler (valores entre 0 y 1)
min_max_scaler = MinMaxScaler()
df[numeric_columns] = min_max_scaler.fit_transform(df[numeric_columns])
df.head(5000)
```

```python
# Comprobación final del tipo de datos
print(df.dtypes)
```
